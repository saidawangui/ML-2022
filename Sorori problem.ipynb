### [Question 1] Number of rice grains on the 100th day

```python
### calculates how many grains of rice you received in total on the 100th day

num_of_days = 100
rice_grain_sum = 1

for i in range(1, num_of_days +1):
    print(i ,rice_grain_sum)
    rice_grain_sum *=2 #multiply each day by 2 in the for loop to get total rice grain
print("Sum of rice in", i, "days is: ", rice_grain_sum)
```

```python
num_days = 100
rice_grain_sum= 1
#creating the values for the x and y axis
yaxis = []
xaxis = list(range(1, num_days + 1))

for i in range(1, num_days + 1,):
    #list the change in the rice grain
    print("For day: {} the sum of grain is: {}".format(i, rice_grain_sum))
    
    #appending the total grain rice into the list of the yaxis
    yaxis.append(rice_grain_sum)
    rice_grain_sum = 2**i    
        
print("Sum of rice in", i, "days is: ", rice_grain_sum)
           
```

```python
#plotting the graph
import matplotlib.pyplot as plt

plt.xlabel('Number of Days')
plt.ylabel('Number of Rice')
plt.title('Change in the Number of Rice Grain per Day')
plt.plot(xaxis, yaxis, linewidth=4.0)
plt.tick_params(labelsize=10)
plt.figure(figsize=(48, 24))
plt.show()
```
